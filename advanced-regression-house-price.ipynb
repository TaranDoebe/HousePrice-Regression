# Cleaning Preperation

```python
def overview_NaN(dataframe):
    
    # Displays amount of NaN values per feature
    missing_values = dataframe.isnull().sum() 

    # Filters columns with no NaN values away
    missing_values = missing_values[missing_values > 0]

    # Formats percentage of missing values
    missing_percentage = (missing_values / len(train_df) * 100).round(1).astype(str) + '%'

    # Create DataFrame for overview
    overview_NaN = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

    return overview_NaN
```

# Feature Engineering

```python
# Feature Engineering
def new_feature_columns(df):
    df['Has_2nd_floor'] = df['2ndFlrSF'] > 0
    df['Has_basement'] = df['BsmtQual'] != 'None'
    df['Has_deductions'] = df['Functional'] != 'Typ'
    df['Has_fireplaces'] = df['Fireplaces'] > 0
    df['Has_garage'] = df['GarageType'] != 'None'
    df['Has_pool'] = df['PoolQC'] != 'None'
    df['Has_fence'] = df['Fence'] != 'None'
    df['Has_WoodDeck'] = df['WoodDeckSF'] > 0
    
    df['Overall_score'] = df['OverallQual'] + df['OverallCond']
    df['Total_building_area'] = df['TotalBsmtSF'] + df["1stFlrSF"] + df["2ndFlrSF"]
    df['TotalBath'] = df['BsmtFullBath'] + 0.5 * df['BsmtHalfBath'] + df['FullBath'] + 0.5 * df['HalfBath']
    df['PorchArea'] = df['OpenPorchSF'] + df['EnclosedPorch'] + df['3SsnPorch'] + df['ScreenPorch']
    
    df['Has_Porch'] = df['PorchArea'] > 0
    
    df['Age'] = df['YrSold'] - df['YearBuilt']
    df['Remodel_age'] = df['YrSold'] - df['YearRemodAdd']
    df['New'] = df['SaleType'] == 'New'
    df['Sold_age'] = 2010 - df['YrSold']
```

# Machine Learning

```python
from sklearn.model_selection import train_test_split, GridSearchCV

X = train_df.drop('SalePrice', axis=1)
y = train_df['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import make_scorer, mean_squared_log_error

def root_mean_squared_log_error(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))

rmsle_scorer = make_scorer(root_mean_squared_log_error, greater_is_better=False)
```

```python
from catboost import CatBoostRegressor
import numpy as np

cat_features = [col for col in X_train.columns if X_train[col].dtype.name == 'category']

catboost_params = {
    'colsample_bylevel': [0.4, 0.5],
    'depth': [4,5,6],
    'iterations': [700,800,900,1000],
    'learning_rate': [0.04],
    'one_hot_max_size': [6,8,10],
    'reg_lambda': [10],
    'subsample': [0.4,0.5]
}

catboost = CatBoostRegressor(
    cat_features=cat_features, 
    random_seed=0, 
    verbose=0
)

catboost_gs = GridSearchCV(
    estimator=catboost, 
    param_grid=catboost_params, 
    cv=5, 
    scoring=rmsle_scorer, 
    n_jobs=-1
)
catboost_gs.fit(X_train, y_train)

catboost_pred = catboost_gs.predict(X_test)
catboost_error = root_mean_squared_log_error(y_test, catboost_pred)

print('CatBoost root mean squared log error: {}'.format(catboost_error))
print('Best parameters: {}'.format(catboost_gs.best_params_))
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

xgb_params = {
    'colsample_bytree': [0.4,0.5],
    'learning_rate': [0.04],
    'max_cat_threshold': [6,8],
    'max_depth': [3,4,5],
    'max_leaves': [8,10,12],
    'n_estimators': [600,700,800,900,1000],
    'reg_alpha': [10],
    'subsample': [0.4,0.5]
}

xgb = XGBRegressor(
    random_state=0,
    enable_categorical=True
)

xgb_gs = GridSearchCV(
    estimator=xgb,
    param_grid=xgb_params,
    cv=5,
    scoring=rmsle_scorer,
    n_jobs=-1
)
xgb_gs.fit(X_train, y_train)

xgb_pred = xgb_gs.predict(X_test)
xgb_error = root_mean_squared_log_error(y_test, xgb_pred)

print('XGBoost root mean squared log error: {}'.format(xgb_error))
print('Best parameters: {}'.format(xgb_gs.best_params_))
```

```python
from sklearn.ensemble import StackingRegressor

stacking_regressor = StackingRegressor(
    estimators=[
        ('catboost', catboost_gs.best_estimator_),
        ('xgboost', xgb_gs.best_estimator_)
    ],
    final_estimator=LinearRegression(),
    n_jobs=-1
)

stacking_regressor.fit(X_train, y_train)

stacking_pred = stacking_regressor.predict(X_test)
stacking_error = root_mean_squared_log_error(y_test, stacking_pred)

print('Stacking root mean squared log error: {}'.format(stacking_error))
```
